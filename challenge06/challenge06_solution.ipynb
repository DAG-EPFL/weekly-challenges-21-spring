## The task

Write a function that finds all the anagrams of a given string. An anagram is a word formed by rearranging the letters of another word (e.g., "LISTEN" and "SILENT").

Here, we do not care whether the generated words are actual dictionary words, we simply want all possible permutations of the string.

The output should simply be the list **(in alphabetical order)** of all possible anagrams. If the string is empty, then simply return the empty string in a (one-element) list.

**Hint:** this task should be feasible without using any library!

```python
#################################################
############# YOUR CODE STARTS HERE #############
#################################################

def find_anagrams(x):
    '''
    Parameters
    --------
    x: string
        The string of which all anagrams want to be computed
    
    Returns
    --------
    anagrams: list[string]
        A list of all possible anagrams sorted alphabetically
    '''
    # Trivial case if the string is not long enough
    return [x] if len(x) < 2 \
        else sorted(list(set( # Sort unique values
            # Every possible letter in x followed by every possible
            # anagram of x without this letter
            x[i] + d
            # For all possible letters in x
            for i in range(len(x))
            # For all possible anagrams of x without this letter
            for d in find_anagrams(x[:i] + x[i+1:])
        )))

#################################################
############## YOUR CODE ENDS HERE ##############
#################################################
```
